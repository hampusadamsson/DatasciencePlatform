```python
import pandas
from tqdm import tqdm
```

```python
df = pandas.read_csv("dataset.csv")
del df["Unnamed: 0"]
df.head()
```

```python
ss = df[["symbol", "open", "date"]]


main_df = pandas.DataFrame()
for s in tqdm(ss.symbol.unique()):
    tmp = ss.loc[ss.symbol == s]
    tmp.set_index("date", inplace=True)
    tmp.rename(columns = {"open": s}, inplace = True)
    tmp.drop(["symbol"], 1, inplace=True)
    if main_df.empty:
        main_df = tmp
    else:
        main_df = main_df.join(tmp, how="outer")
```

```python
df2 = main_df.fillna(0)
delta = (df2.shift(1) - df2) / df2
delta
```
